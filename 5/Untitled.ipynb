```python
>>> re.findall( r'^.*(?:ing|ly|ed|ious|ies|ive|es|s|ment)$', 'singing' )
```

```python
>>> re.findall( r'^(.*?)(ing|ly|ed|ious|ies|ive|es|s|ment)$', 'apple' )
```

```python
>>> re.findall(r'^(.*?)(ing|ly|ed|ious|ies|ive|es|s|ment)?$' , 'apple' )
```

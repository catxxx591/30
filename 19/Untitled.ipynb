```python
import nltk
```

```python
stopwords=[]
with open('stopwords_ch.txt', 'r', encoding='UTF-8') as file:
    for data in file.readlines():
        data = data.strip()
        stopwords.append(data)
```

```python
lab_tags ={}
#txt =''
prelab = ''
books = list(news.items())
for book in books:

#    n = 0
    for lab in ['即時','生活','運動','政治','科技','國際','專題','兩岸','產經','社會','地方','娛樂','證券','文化']:
        if book[1]==lab:
            if prelab != lab:
                txt=''
            txt = txt+book[0]
            lab_tags[book[1]] = txt
            prelab = lab
#        n += 1
    
```

```python
filterwords=[
#    '年','月','中央社','記者','編輯',
    '\n',' ','「','」'
    ,'今天','說','表示'
]
```

```python
lab_fwords={}
for labname in lab_tags.keys():
    print(labname,end=',')
    lab_words = lab_tags[labname]
    seg_list = jieba.cut(lab_words, cut_all=False)
    txt = ",".join(seg_list)
    tokens = txt.split(',')
    tokens = list(filter(lambda a: a not in stopwords and a not in filterwords, tokens))
    lab_fwords[labname] = nltk.FreqDist(tokens).most_common(2000)
```

```python
import random
```

```python
def get_featuresets(books):
    features = {}
    featuresets=[]
    for book in books:
        txt = book[0]
        labname = book[1]
        seg_list = jieba.cut(txt, cut_all=False)
        txt = ",".join(seg_list)
        tokens = txt.split(',')
        for tok in tokens:
            features[tok] = tok in [i[0] for i in lab_fwords[labname]]
        featuresets.append((features,labname))
    random.shuffle(featuresets)
    return featuresets
```

```python
featuresets = get_featuresets(books)
```

```python
len(featuresets)
```

```python
train_set, test_set = featuresets[:840], featuresets[840:]
classifier = nltk.NaiveBayesClassifier.train(train_set)
nltk.classify.accuracy(classifier, test_set)
```

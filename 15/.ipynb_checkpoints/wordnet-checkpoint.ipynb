```python
import nltk
>>> from nltk.corpus import movie_reviews
>>> import random
import re
from nltk.corpus import wordnet as wn
```

```python
>>> #nltk.download()
```

```python
>>> documents = [(list(movie_reviews.words(fileid)), category)
...              for category in movie_reviews.categories()
...              for fileid in movie_reviews.fileids(category )]
>>> random.shuffle(documents)
```

```python
book_syns = []
for book in documents:
    booktxt = book[0]
    lab = book[1]
    book_syns.append( ([wn.synsets(w)[0] for w in booktxt if len(wn.synsets(w)) > 0],lab))

book_syns[:10]
```

```python
book_syns[0:10]
```

```python
pos = wn.synsets('positive')[0]
neg = wn.synsets('negative')[0]
```

```python
featuresets = []
for syns,lab in book_syns:
    feature = {}
    for syn in syns:
        word = syn.lemmas()[0].name()
        pos_value = pos.path_similarity(syn)
        neg_value = neg.path_similarity(syn)
        if pos_value and neg_value != None:
            feature[word]=pos_value-neg_value
    featuresets.append((feature,lab))
```

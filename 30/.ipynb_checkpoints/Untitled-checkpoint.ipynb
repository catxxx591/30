```python
from zhtools.langconv import *
def simple2tradition(sent):
    #将简体转换成繁体
    w=''
    for line in sent:
        line = Converter('zh-hant').convert(line)#.decode('utf-8'))
        line = line#.encode('utf-8')
        w += line
    return w

def tradition2simple(line):
    # 将繁体转换成简体
    line = Converter('zh-hans').convert(line)#.decode('utf-8'))
    line = line.encode('utf-8')
    return line
```

```python
stop_words = [re.findall(r'\S+',w)[0] for w in open('../txt/stopwords_ch.txt',encoding='utf8').readlines() if len(re.findall(r'\S+',w))>0]
def sent2word(sentence,stop_words=stop_words):
    words = jieba.cut(sentence, HMM=False)
    words = [w for w in words if w not in stop_words]
    return words
```

```python
def get_sentment(sent):
    tokens = sent2word(sent)
    score = 0
    countword = 0
    for w in tokens:
        
        if w in sentment_dict.keys():
 #           print((w,sentment_dict[w]))
            score += sentment_dict[w]
            countword += 1
    if countword != 0:
        return score/countword
    else:
        return 0
```

```python
sent ="我覺得海豚非常聰明"

get_sentment(sent)
```

```python
def pos_feature(sent):
    i = sent
    feature={}
#    words = i[0].split(" ")
    words = sent2word(sent)
    for w in words:
        if w in sentment_dict.keys():
            feature[w] = sentment_dict[w]   
#        feature[w] = w in pos_dict
    return feature
```

```python
def get_featureset(wb30):
    featureset = []
    for i in wb30:
        sent = ''
        for w in i[0].split():
            sent += w
#        print(sent)
        feature = pos_feature(sent)
#for i in wb30:
#    feature={}
#    words = i[0].split(" ")
#    for w in words:
#        if w in sentment_dict.keys():
#            feature[w] = sentment_dict[w]
        if  len(feature.keys()) > 0:
#            print(feature[w])
            featureset.append((feature,i[1]))
    return featureset
```

```python
sent='這支手機很漂亮'
print(len(wb30))
featuresets = get_featureset(wb30)
#featuresets
#sent2word(sent)
train_set, test_set = featuresets[:20000], featuresets[20000:]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier, test_set))
```

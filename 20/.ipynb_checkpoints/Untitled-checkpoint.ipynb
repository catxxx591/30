```python
labs_soup=[]
for item in labs:
    soups=[]
    n=1
    while n < 13:
        url = 'https://www.cna.com.tw/cna2018api/api/simplelist/categorycode/'+item[0]+'/pageidx/'+str(n)
        res = requests.get(url)
        soup = BeautifulSoup(res.text,'html5lib')
        body = soup.findAll('body')[0].getText()
        PageUrl = re.findall(r'\"PageUrl\"\:\S[^\,]+',body)
        urls = [txt[11:-1] for txt in PageUrl]
        soups.extend(urls)
        n += 1
        if len(urls) < 20:
            break
    labs_soup.append([soups,item[1]])
```

```python
lab_tags ={}
#txt =''
prelab = ''
books = list(news.items())
for book in books:

#    n = 0
    for lab in ['生活','運動','政治','科技','國際','兩岸','產經','社會','地方','娛樂','證券','文化']:
        if book[1]==lab:
            if prelab != lab:
                txt=''
            txt = txt+book[0]
            lab_tags[book[1]] = txt
            prelab = lab
#        n += 1
    
```

```python
filterwords=[
    '年','月','中央社','記者','編輯','譯者','日電',
    '\n',' ','「','」'
    ,'今天','說','表示','年','月','日','後','前','時','應','中','今年'
    ,'兩','下','人','最','次','再','先'
]
```

```python
lab_fwords={}
for labname in lab_tags.keys():
    print(labname,end=',')
    lab_words = lab_tags[labname]
    seg_list = jieba.cut(lab_words, cut_all=False)
    txt = ",".join(seg_list)
    tokens = txt.split(',')
    tokens = list(filter(lambda a: a not in stopwords and a not in filterwords ,tokens))#and len(a)>1, tokens))
    lab_fwords[labname] = nltk.FreqDist(tokens).most_common(200)
```

```python
def pos_features(article,top):
    features={}
    t = ''
    seg_list = jieba.cut(article, cut_all=False)
    t = ",".join(seg_list)
    article_tokens = t.split(',')
    filt_tokens = list(filter(lambda a: a not in stopwords and a not in filterwords , article_tokens))#and len(a)>1, article_tokens))
    toks = [w[0] for w in nltk.FreqDist(filt_tokens).most_common(top)]
#   print(toks)
    for tok in toks:
#            features[tok] = tok in [i[0] for i in lab_fwords[labname]]
        features[tok] = True
    return features
```

```python
def get_featuresets(books,top=200):

    featuresets=[]
    for book in books:
#        features = {}
        article = book[0]
        labname = book[1]
        features = pos_features(article,top)
        feature_item = (features,labname)
        featuresets.append(feature_item)

        print(labname)

    return featuresets
```

```python
featuresets = get_featuresets(books,200)
```

```python
size = int(len(featuresets)*0.7)
train_set, test_set = featuresets[:size], featuresets[size:]
classifier = nltk.NaiveBayesClassifier.train(train_set)
nltk.classify.accuracy(classifier, test_set)
```
